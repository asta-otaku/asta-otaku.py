```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("data.csv")
```

```python
data.info()
```

```python
data
```

```python
#Best Players
data.sort_values(['Overall'],ascending=False).head(10)
```

```python
#Total Number of right and left foot players
data["Preferred Foot"].value_counts()
```

```python
sns.set_style('darkgrid')
plt.figure(figsize=(8,5))
sns.countplot(x= data['Preferred Foot'],palette=sns.color_palette('winter_r'),edgecolor=('black'))
plt.title('Preffered Foot',size=25)
plt.xlabel('Foot',size=15)
plt.ylabel('Count',size=15)
```

```python
#Distribution of player nationalities
data.groupby("Nationality").count()["ID"]
```

```python
nation = data['Nationality'].value_counts().reset_index()
nation.columns = ['Country', 'No of Players']
nation = nation.set_index('Country').sort_values(['No of Players'],ascending= False)
plt.figure(figsize=(20,8))
sns.barplot(x=nation.index,y=nation['No of Players'],palette='Spectral',edgecolor='black')
plt.xticks(rotation='vertical',size=8)
plt.title('Players per Nation',size=25)
plt.ylabel('Players Count',size=15)
plt.xlabel('Country',size=15)
plt.show()
```

```python
data["Wage"] = data["Wage"].str.replace('€',"")
```

```python
data["Wage"] = data["Wage"].str.replace('K',"000")
```

```python
data["Wage"] = data["Wage"].astype('float')
```

```python
#Player Wages
data[['Name','Wage']]
```

```python
#Average Rating of Players
data[["Name","Overall"]]
```

```python
#Top 10 Players according to Potential
data.sort_values(['Potential'],ascending=False)['Name'].head(10)
```
